```python
# Assuming 'traits_df' is your DataFrame and 'highScore' is the column you want to keep as y
# Melt the DataFrame into a long format
trait_scores_df = get_trait_scores_df(leaderboard_df, traits_df)
trait_scores_df
```

```python
traits_df.describe()
```

```python

# Create the FacetGrid
g = get_trait_score_facet_grid(trait_scores_df)

# Map a scatter plot to the FacetGrid
g.map(plt.scatter, "value", "highScore", alpha=0.2)

# Set the maximum number of x ticks for each facet
for ax in g.axes.flat:
    ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
```

```python
# Create the FacetGrid
g = get_trait_score_facet_grid(trait_scores_df)

# Map a scatter plot to the FacetGrid
g.map(sns.histplot, "value", kde=True)

# Set the maximum number of x ticks for each facet
for ax in g.axes.flat:
    ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
```

```python

# Create the FacetGrid
g = get_trait_score_facet_grid(trait_scores_df)

# Map a kde plot to the FacetGrid
g.map(sns.kdeplot, "value", "highScore", cmap='hot', fill=True)

# Set the maximum number of x ticks for each facet
for ax in g.axes.flat:
    ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
```
