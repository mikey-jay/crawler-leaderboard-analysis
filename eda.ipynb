```python
gotchi_ids = leaderboard_df.sort_values('highScore', ascending=False).head(1000).index.to_list()
gotchis_df = get_gotchis_df(gotchi_ids)
gotchis_df.head()
```

```python
traits_df = gotchis_df['modifiedNumericTraits'].apply(pd.Series).rename(columns={0:'NRG', 1:'AGG', 2:'SPK', 3:'BRN', 4:'EYS', 5:'EYC'}).merge(gotchis_df[['kinship', 'withSetsRarityScore']], left_index=True, right_index=True)
traits_df.head()
```

```python
traits_df.info()
```

```python
traits_df.corrwith(leaderboard_df['highScore']).sort_values(key=abs, ascending=False)
```

```python
# Assuming 'traits_df' is your DataFrame and 'highScore' is the column you want to keep as y
# Melt the DataFrame into a long format
long_df = pd.concat([traits_df, leaderboard_df['highScore']], axis='columns').melt(id_vars='highScore', var_name='trait', value_name='value')
long_df
```

```python
traits_df.describe()
```

```python

# Create the FacetGrid
g = sns.FacetGrid(long_df, col="trait", sharex=False, col_wrap=4, height=4)

# Map a scatter plot to the FacetGrid
g.map(plt.scatter, "value", "highScore", alpha=0.2)

# Set the maximum number of x ticks for each facet
for ax in g.axes.flat:
    ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
```

```python
# Create the FacetGrid
g = sns.FacetGrid(long_df, col="trait", sharex=False, col_wrap=4, height=4)

# Map a scatter plot to the FacetGrid
g.map(sns.histplot, "value", kde=True)

# Set the maximum number of x ticks for each facet
for ax in g.axes.flat:
    ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
```

```python

# Create the FacetGrid
g = sns.FacetGrid(long_df.query('highScore > 5000'), col="trait", sharex=False, col_wrap=4, height=4)

# Map a kde plot to the FacetGrid
g.map(sns.kdeplot, "value", "highScore", cmap='hot', fill=True)

# Set the maximum number of x ticks for each facet
for ax in g.axes.flat:
    ax.xaxis.set_major_locator(ticker.MaxNLocator(nbins=5))
```
